**Question 2:**

a. **Classification:** This algorithm identifies groups in which a data element can belong to. It takes the number of nearest neighbors to classify the elements. It is a supervised ML as initial data points are necessary.
Example: Data abstraction in federal organizations that dictates the access limitations of each employee.<br>
b. **Clustering:** This algorithm converts the dataset into meaningful groups of objects usually based on a distance metric. It is an unsupervised ML algorithm as an explicit outcome isn’t defined.
Example: Genetic Taxonomy of all living things into species, kingdom, phylum, class, order, family etc. <br>
c. **Density estimation:** Its another algorithm to construct a density estimate based on a dataset by means of its underlying probability density function. It is an unsupervised learning task.
Example: The density estimates for women who have diabetes given the levels of their plasma glucose.<br>
d. **Regression:** It is an algorithm used to predict/analyze the impact of independent variables on the dependent variables. It is a supervised ML algorithm as we need to provide some examples of known outcomes to build the learning model. 
Example: Prediction of snowfall this year based on past decade, Stock market predictions based on the history of the firms.<br>
e. **Samples:** They are individual data points representative of the entire datasets, used to analyze trends. 
Example: A sample (25 children) of the child population aged 2-5 years, is used to study cavity growth due to candy consumption.<br> 
f. **Features:** It is used to describe a large set of data points.
Example: The data columns like UserID in twitter feed analysis.<br>
g. **Multivariate data:** Describes data that has more than one variable. It is a data set that mimics reality.
Example: The several factors that affect market predictions for stock prices.<br>
h. **Training set:** The data set used to make predictions. 
Example: Correlation between bar sales and temperature. The prediction of sales given a temperature. In this case, the algorithm is supervised.<br>
i. **Test set:** The dataset used to test the prediction. 
Example: Proof that bar sales and the temperature are linearly related. <br>

```python
print(__doc__)

# Author: Gael Varoquaux <gael dot varoquaux at normalesup dot org>
# License: BSD 3 clause

# Standard scientific Python imports
import matplotlib.pyplot as plt                           #Imports matplotlib package

from sklearn import datasets, svm, metrics                # Import datasets, classifiers and performance metrics from sklearn

# The digits dataset
digits = datasets.load_digits()

# The data that we are interested in is made of 8x8 images of digits, let's
# have a look at the first 4 images, stored in the `images` attribute of the
# dataset.  If we were working from image files, we could load them using
# matplotlib.pyplot.imread.  Note that each image must have the same size. For these
# images, we know which digit they represent: it is given in the 'target' of
# the dataset.
images_and_labels = list(zip(digits.images, digits.target))   # Combines two lists using zip function for handling ease.
for index, (image, label) in enumerate(images_and_labels[:4]): # The first 4 items in the data set are plotted.
    plt.subplot(2, 4, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Training: %i' % label)

# To apply a classifier on this data, we need to flatten the image, to
# turn the data in a (samples, feature) matrix:
n_samples = len(digits.images)                           # the 8x8 matrix needs to become a flat array of 64 pixels 
data = digits.images.reshape((n_samples, -1))              

# Create a classifier: a support vector classifier
classifier = svm.SVC(gamma=0.001)

# We learn the digits on the first half of the 'digits'
classifier.fit(data[:n_samples // 2], digits.target[:n_samples // 2])

# Now predict the value of the digit on the second half:
expected = digits.target[n_samples // 2:]
predicted = classifier.predict(data[n_samples // 2:])

print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(expected, predicted)))
print("Confusion matrix:\n%s" % metrics.confusion_matrix(expected, predicted))

images_and_predictions = list(zip(digits.images[n_samples // 2:], predicted)) # If satisfied with above results, the machine can be used to predict some images.. 
for index, (image, prediction) in enumerate(images_and_predictions[:4]):
    plt.subplot(2, 4, index + 5)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Prediction: %i' % prediction)

#plt.show()
```

The function of the above code block is to train the machine to identify arbitrary set of images of a digit. Matplotlib is used for the visualization of the images while the fit function is used to train the machine with the training dataset. The output of the classifier is precision, recall, f1-score, support metrics for the corresponding digit targets and a confusion matrix. The precision of the output will vary if the training dataset varies in size.
